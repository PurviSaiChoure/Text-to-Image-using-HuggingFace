```python
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
import torch
```

```python
model_id1 = "dreamlike-art/dreamlike-diffusion-1.0"
model_id2 = "stabilityai/stable-diffusion-xl-base-1.0"

pipe = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16, use_safetensors=True)
pipe = pipe.to("cuda")
```

```python
prompt = """fantastical art, grunge woman with vibrant neon hair, mid-step through a swirling dimension, joyful expression, large soft eyes, heart-shaped face, high contrast lighting, slim frame, wild hair cascading down, distressed anime shirt and loose-fitting denim
"""
```

```python
image = pipe(prompt).images[0]
```

```python
print("[PROMPT]: ",prompt)
plt.imshow(image);
plt.axis('off');
```

```python
prompt2 = """psychedelic vibe, a mystical girl with rainbow-colored hair, reaching out between portals, serene eyes, pointed chin, intense bokeh lights, delicate frame, sleek straight hair, distressed kawaii crop top with relaxed-fit jeans
"""
image = pipe(prompt2).images[0]
```

```python
print('[PROMPT]: ',prompt2)
plt.imshow(image);
plt.axis('off');
```
